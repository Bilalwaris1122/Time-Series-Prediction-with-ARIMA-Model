```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
def arima_objective_function(trial: Trial, data_values):
    p = trial.suggest_int('p', 0, 11)

    q = trial.suggest_int('q', 0, 11)
    trend = trial.suggest_categorical('trend', ['n', 'c', 't', 'ct'])

    try:
        model = ARIMA(data_values, order=(p, 0, q), trend=trend)  # Setting d and D to 0
        predictions = model.fit()
        mse = mean_squared_error(data_values, predictions.fittedvalues)
    except:
        mse = 1e5

    return mse
```
